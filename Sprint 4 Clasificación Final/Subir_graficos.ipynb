```python
import requests
from bs4 import BeautifulSoup

# URL de la carpeta donde están los CSV en GitHub
github_folder_url = "https://github.com/metaute73/Proyecto-Gen-mica-2024-2/tree/main/Ensamblajes"

# Obtener el HTML de la página
response = requests.get(github_folder_url)
soup = BeautifulSoup(response.text, 'html.parser')

# Buscar enlaces que terminan en ".csv"
csv_files = [a.text for a in soup.find_all('a', href=True) if a.text.endswith('.csv')]

print("Archivos CSV encontrados:")
print(csv_files)
```

```python
import os

# Carpeta donde guardar los CSV en Colab
csv_folder = "/content/Ensamblajes"
os.makedirs(csv_folder, exist_ok=True)

# Descargar cada CSV
base_url = "https://raw.githubusercontent.com/metaute73/Proyecto-Gen-mica-2024-2/main/Ensamblajes/"

for file in csv_files:
    file_url = base_url + file
    file_path = os.path.join(csv_folder, file)

    response = requests.get(file_url)

    if response.status_code == 200:
        with open(file_path, 'wb') as f:
            f.write(response.content)
        print(f"Descargado: {file}")
    else:
        print(f"Error al descargar: {file}")
```

```python
import os
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

def detect_delimiter(file_path):
    """Detecta automáticamente el delimitador del archivo (tabulación o coma)."""
    with open(file_path, 'r') as file:
        first_line = file.readline()
        return "," if "," in first_line else "\t"

def process_genomic_data(csv_file, output_dir="graficos"):
    """
    Carga un archivo CSV, filtra los datos, elimina valores atípicos usando MAD y genera visualizaciones.

    Se generan 5 gráficos organizados en una subcarpeta (por experimento):
      1. Histograma de 'Length'
      2. Histograma de 'GC'
      3. Histograma de 'Depth'
      4. Histograma de 'AltAllels'
      5. Scatter plot de 'Length' vs 'Depth'

    Parámetros:
        csv_file (str): Ruta del archivo CSV.
        output_dir (str): Carpeta donde se guardarán las subcarpetas de gráficos.

    Retorna:
        pd.DataFrame: DataFrame procesado.
    """
    # Asegurar que la carpeta principal de salida exista
    os.makedirs(output_dir, exist_ok=True)

    # Detectar delimitador y cargar datos
    delimiter = detect_delimiter(csv_file)
    df = pd.read_csv(csv_file, delimiter=delimiter)

    # Filtrar por longitud mínima
    df = df[df['Length'] >= 1000]

    # Eliminar valores atípicos en 'Depth' usando el método MAD
    median_depth = df['Depth'].median()
    MAD = np.median(np.abs(df['Depth'] - median_depth))
    threshold_mad = 3 * MAD
    df = df[(df['Depth'] >= median_depth - threshold_mad) & (df['Depth'] <= median_depth + threshold_mad)]

    # Crear una subcarpeta para el experimento usando el nombre del archivo
    base_name = os.path.basename(csv_file).replace(".csv", "")
    experiment_folder = os.path.join(output_dir, base_name)
    os.makedirs(experiment_folder, exist_ok=True)

    # Gráfico 1: Histograma de Length
    plt.figure(figsize=(8, 5))
    sns.histplot(df["Length"], bins=20, kde=True, color="seagreen")
    plt.title("Distribución de la Longitud de los Scaffolds")
    plt.tight_layout()
    plt.savefig(os.path.join(experiment_folder, f"{base_name}_Length.png"))
    plt.close()

    # Gráfico 2: Histograma de GC
    plt.figure(figsize=(8, 5))
    sns.histplot(df["GC"], bins=20, kde=True)
    plt.title("Distribución del Contenido GC")
    plt.tight_layout()
    plt.savefig(os.path.join(experiment_folder, f"{base_name}_GC.png"))
    plt.close()

    # Gráfico 3: Histograma de Depth
    plt.figure(figsize=(8, 5))
    sns.histplot(df["Depth"], bins=20, kde=True, color="orange")
    plt.title("Distribución de la Profundidad de los Scaffolds")
    plt.tight_layout()
    plt.savefig(os.path.join(experiment_folder, f"{base_name}_Depth.png"))
    plt.close()

    # Gráfico 4: Histograma de AltAllels
    plt.figure(figsize=(8, 5))
    sns.histplot(df["AltAllels"], bins=10, kde=True, color="cyan")
    plt.title("Distribución de los Alelos Alternativos")
    plt.tight_layout()
    plt.savefig(os.path.join(experiment_folder, f"{base_name}_AltAllels.png"))
    plt.close()

    # Gráfico 5: Scatter plot Length vs Depth
    plt.figure(figsize=(8, 6))
    sns.scatterplot(x='Depth', y='Length', data=df)
    plt.xlabel("Depth")
    plt.ylabel("Length")
    plt.title("Length vs Depth")
    plt.tight_layout()
    plt.savefig(os.path.join(experiment_folder, f"{base_name}_length_vs_depth.png"))
    plt.close()

    return df
```

```python
import glob

csv_files = glob.glob("Ensamblajes/*.csv")

for file in csv_files:
    process_genomic_data(file)

print("✅ ¡Todos los gráficos han sido generados y guardados!")
```
